```python
from laliga_players_by_team import data
import pandas as pd
```

```python
teamData = data.copy()
```

```python
def cleanup():
    data2 = []
    data3 = []
    data = teamData[2::3]
    for i in data:
        data2.append(i.strip())

    for i in data2:
        data3.append(i.split('\n'))

    #can be modified depending on size of team
    data3 = data3[:26]
    

    return data3

data3 = cleanup()
```

```python
df = pd.DataFrame(data3, columns=['number', 'name', 'position', 'age', 'market_value'])
df = df.drop(25)
```

```python
for index in df.itertuples():
    if len(index[4].split('€')) == 1:
        df['age'][index[0]] = index[4].split('€')[0]

    if len(index[4].split('€')) > 1:
        df.market_value[index[0]] = index[4].split('€')[1]
        df['age'][index[0]] = index[4].split('€')[0]
```

```python
df.market_value = df.market_value.str.replace('€', '').str.replace('Th.', '').str.replace('m', '')
df.market_value = pd.to_numeric(df.market_value, downcast='float')
df.market_value[2] = 0.4
```

```python
df['birthday'] = None

for index in df.itertuples():
    df.birthday[index[0]] = index[4].split('(')[0]

for index in df.itertuples():
    df.age[index[0]] = index[4].split('(')[1].replace(')', '')

df.birthday = pd.to_datetime(df.birthday)
```
